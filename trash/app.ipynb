```python
# check if file is csv or excel
def is_csv(filename):
    # Get the file extension from the filename
    file_extension = os.path.splitext(filename)[1].lower()

    # Check if the file extension is CSV or Excel
    if file_extension in ['.csv']:
        return True
    else:
        return False
# check if file is csv or excel
```
